```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import  requests as req
import pandas as pd
import time
```

```python
url = "https://www.imf.org/external/np/fin/tad/balmov2.aspx?type=TOTAL"
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
# time.sleep(2)
# ---------- ############# ------------- #
driver.get(url)
time.sleep(3)
# ---------- ############# ------------- #
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
# ---------- ############# ------------- #
# print(soup.prettify())

table = soup.find("table", {"id": "gvMovements"})
print(f"Table tag : {table}")

table_rows = table.find_all("tr", {"style" : "height:25px;"})
print(f"table rows tags : {table_rows}")

Members = []
Credit_Outstanding = []
Total_Disbursements = []
Total_Repayments = []
Total_IMF_Credit_Outstanding = []

for data in table_rows:
    # member = data.find("tr", {"style" : "height:25px;"}).find("td").get_text(strip = True)
    # credit_outstanding = int(data.find("tr", {"style" : "height:25px;"}).find("td", {"align" : "right"}))[1]
    # total_disbursements = int(data.find("tr", {"style" : "height:25px;"}).find("td", {"align" : "right"}))[2]
    # total_repayments = int(data.find("tr", {"style" : "height:25px;"}).find("td", {"align" : "right"}))[3]
    # total_imf_credit_outstanding = int(data.find("tr", {"style" : "height:25px;"}).find("td", {"align" : "right"}))[5]

    cols = data.find_all("td")   # get all columns

    if len(cols) >= 5:   # avoid errors
        member = cols[0].get_text(strip=True)
        credit_outstanding = cols[1].get_text(strip=True)
        total_disbursements = cols[2].get_text(strip=True)
        total_repayments = cols[3].get_text(strip=True)
        total_imf_credit_outstanding = cols[4].get_text(strip=True)

        # append to list
        Members.append(member)
        Credit_Outstanding.append(credit_outstanding)
        Total_Disbursements.append(total_disbursements)
        Total_Repayments.append(total_repayments)
        Total_IMF_Credit_Outstanding.append(total_imf_credit_outstanding)
# ---------- ############# ------------- #
driver.quit()
```

```python
print(Members, Credit_Outstanding, Total_Disbursements, Total_Repayments, Total_IMF_Credit_Outstanding)
```

```python
data = {
    "Members" : Members,
    "Total_IMF_Credit_Outstanding_as_of_01/31/2026" : Credit_Outstanding,
    "Total_Disbursements" : Total_Disbursements,
    "Total_Repayments" : Total_Repayments,
    "Total_IMF_Credit_Outstanding_as_of_02/20/2026" : Total_IMF_Credit_Outstanding
}
```

```python
df = pd.DataFrame(data)
```
